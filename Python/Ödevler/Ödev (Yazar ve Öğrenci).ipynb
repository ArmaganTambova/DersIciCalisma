--
En az 20 adet öğrencinin isim soyad şehir ve ilçe (dize) bilgileri kayıtlı olacak bir menü listelenecek
>- 1 öğrenci ekle
>- 2 öğrenci bul
Eğer ki öğrenci ekle seçilirse öğrencinin isim soyad şehir ve ilçe bilgileri teker teker alınıp dizilere ekelenecek
Eğer 2 seçilirse yeni bir menü listelenecek 
>- 1 şehire göre
>- 2 ilçeye göre

Seçilen menü elemanına göre arama yapılıp öğrenciler listeleneccek
Not:
Tüm bu işlemler klavyede "h" tuşuna basılana kadar tekrar tekrar devam edilecek.

```python
isimler = ["Armağan", "Nisa Nur", "Cemrenaz", "Ecrin Zümra", "Alperen Efe", "Alperen", "Ceyhun", "Onur", "Muhammed Yusuf", "İbrahim", "Ekrem İbrahim", "Yüksel", "Melike Zeynep", "Burak", "Bilge", "Naz", "Selim", "Kaan", "Deniz", "Doğa"]
soyadlar = ["Tambova", "Uçar", "Türkanoğlu", "Koparan", "Düz", "Fidan", "Topçu", "Girgin", "Uşun", "Savaş", "Azizoğlu", "Girgin", "Öz", "Uzun", "Poyraz", "Sağım", "Çelik", "Aydıngör", "Aytüre", "Çopur"]
sehirler = ["Eskişehir", "Eskişehir", "Ankara", "Eskişehir", "Ankara", "Eskişehir", "Eskişehir", "Eskişehir", "Eskişehir", "Eskişehir", "Eskişehir", "Eskişehir", "İstanbul", "Eskişehir", "Eskişehir", "Eskişehir", "Eskişehir", "Eskişehir", "Eskişehir", "Eskişehir"]
ilceler = ["Tepebaşı", "Tepebaşı", "Çankaya", "Tepebaşı", "Çankaya", "Tepebaşı", "Tepebaşı", "Tepebaşı", "Tepebaşı", "Tepebaşı", "Tepebaşı", "Tepebaşı", "Beşiktaş", "Odunpazarı", "Odunpazarı", "Tepebaşı", "Odunpazarı", "Tepebaşı", "Tepebaşı", "Tepebaşı"]

secenekend = "Q"

print("1) Öğrenci Ekle")
print("2) Öğrenci Bul")
secim = input("Lütfen Seçim Yapınız")
while secenekend != "q":
    while secim == "1":
        isim = input("Lütfen Kayıt Etmek İstediğiniz Öğrencinin Adını Giriniz. ÖRNEK: Armağan")
        isimler.append(isim)
        soyad = input("Lütfen " + isim + " Adlı Öğrencinin Soyadını Giriniz. ÖRNEK: Tambova")
        soyadlar.append(soyad)
        sehir = input("Lütfen " + isim + " " + soyad + " Adlı Ve Soyadlı Öğrencinin Kayıtlı Olduğu Şehri Giriniz. ÖRNEK: Eskişehir")
        sehirler.append(sehir)
        ilce = input("Lütfen " + isim + " " + soyad + " Adlı Ve Soyadlı Öğrencinin Kayıtlı Olduğu İlçeyi Giriniz. ÖRNEK: Tepebaşı")
        ilceler.append(ilce)
        print("Öğrenci Kayıtınız Tamamlanmıştır")
        secenekkdevam = input("Kayıt İşlemine Devam Etmek İstiyor Musunuz? e/h")
        if secenekkdevam == "e":
            secenekend = "Q"
            secim = "1"
        elif secenekkdevam == "h":
            secim1 = input("Öğrenci Bulmak İster Misiniz? e/h")
            if secim1 == "e":
                secenekend = "Q"
                secim = "2"
            elif secim1 == "h":
                print("İyi Günler Dileriz. İşlemleriniz Tamamlandı")
                secenekend = "q"
                secim = "Q"
            else:
                print("Geçerli Bir Seçenek Girmediniz. Lütfen Daha Sonra Tekrar Deneyiniz")
                secenekend = "q"
        else:
            print("Geçerli Bir Seçenek Girmediniz. Lütfen Daha Sonra Tekrar Deneyiniz")
            secenekend = "q"
            secim = "Q"
    while secim == "2":
        print("1) İle Göre Ara")
        print("1) İlçeye Göre Ara")
        secim2 = input("Lütfen Seçim Yapınız")
        if secim2 == "1":
            sehirsecim = input("Lütfen İli Giriniz. ÖRNEK: Eskişehir")
            sayac = 0
            kontrol = 0
            print(sehirsecim + " İline Kayıtlı Olan Öğrenciler")
            while sayac < len(sehirler):
                if sehirsecim == sehirler[sayac]:
                    print(isimler[sayac] + " " + soyadlar[sayac] + " " + ilceler[sayac])
                else:
                    kontrol += 1
                sayac += 1
            if kontrol == len(sehirler):
                print(sehirsecim + " İline Kayıtlı Öğrenci Bulunamamıştır")
            secenekbdevam = input("Öğrenci Bulmaya Devam Etmek İster Misiniz? e/h")
            if secenekbdevam == "e":
                secenekend = "Q"
                secim = "2"
            elif secenekbdevam == "h":
                secim3 = input("Öğrenci Eklemek İster Misiniz? e/h")
                if secim3 == "e":
                    secenekend = "Q"
                    secim = "1"
                elif secim3 == "h":
                    print("İyi Günler Dileriz. İşlemleriniz Tamamlandı")
                    secenekend = "q"
                    secim = "Q"
                else:
                    print("Geçerli Bir Seçenek Girmediniz. Lütfen Daha Sonra Tekrar Deneyiniz")
                    secenekend = "q"
                    secim = "Q"
            else:
                print("Geçerli Bir Seçenek Girmediniz. Lütfen Daha Sonra Tekrar Deneyiniz")
                secenekend = "q"
                secim = "Q"
        if secim2 == "2":
            ilcesecim = input("Lütfen İlçeyi Giriniz. ÖRNEK: Odunpazarı")
            sayac = 0
            kontrol = 0
            print(ilcesecim + " İlçesine Kayıtlı Öğrenciler")
            while sayac < len(ilceler):
                if ilcesecim == ilceler[sayac]:
                    print(isimler[sayac] + " " + soyadlar[sayac] + " " + sehirler[sayac])
                else:
                    kontrol += 1
                sayac += 1
            if kontrol == len(ilceler):
                print(ilcesecim + " İlçesine Kayıtlı Öğrenci Bulunamamıştır")
            secenekbdevam = input("Öğrenci Bulmaya Devam Etmek İster Misiniz? e/h")
            if secenekbdevam == "e":
                secenekend = "Q"
                secim = "2"
            if secenekbdevam == "h":
                secim3 = input("Öğrenci Eklemek İster Misiniz? e/h")
                if secim3 == "e":
                    secenekend = "Q"
                    secim = "1"
                elif secim3 == "h":
                    print("İyi Günler Dileriz. İşlemleriniz Tamamlandı")
                    secenekend = "q"
                    secim = "Q"
                else:
                    print("Geçerli Bir Seçenek Girmediniz. Lütfen Daha Sonra Tekrar Deneyiniz")
                    secenekend = "q"
                    secim = "Q"
            else:
                print("Geçerli Bir Seçenek Girmediniz. Lütfen Daha Sonra Tekrar Deneyiniz")
                secenekend = "q"
                secim = "Q"
```
